```python
def contar_palavras(palavras_ordenadas):
    pa_ordenada = sorted(palavras_ordenadas)
    palavra_anterior = pa_ordenada[0]
    contagem = 1
    
    for palavr in pa_ordenada[1:]:
        if palavr == palavra_anterior:
            contagem += 1
        else:
            print(f'{palavra_anterior}: {contagem}')
            palavra_anterior = palavr
            contagem = 1
    print(f'{palavra_anterior}: {contagem}')
            
if __name__ == '__main__':   
    contar_palavras(palavras_ordenadas)
    print()
```
